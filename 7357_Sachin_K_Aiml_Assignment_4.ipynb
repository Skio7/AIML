```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
diabetes_df = pd.read_csv("diabetes.csv")
diabetes_data = diabetes_df[["148","72","35"]]
diabetes_data.to_csv("cluster_data.csv",index=False)
```

```python
filtered_df = pd.read_csv("cluster_data.csv")
filtered_df.head()
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(filtered_df)

labels = kmeans.predict(filtered_df)
centroids = kmeans.cluster_centers_

filtered_df['Clusters'] = labels
filtered_df.head()
# print(centroids)
```

```python
import matplotlib.pyplot as plt

#For creating a new plot
fig = plt.figure()

fig = plt.figure(figsize=(10, 6)) 
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

# Scatter plot using col_2, col_3, and col_4
ax.scatter(filtered_df['148'], filtered_df['72'], filtered_df['35'], c=filtered_df['Clusters'], cmap='rainbow')

ax.set_xlabel('col_2(148)')
ax.set_ylabel('col_3(72)')
ax.set_zlabel('col_4(35)')
ax.set_title('K-Means Clustering (3D Scatter Plot)')
```
